```python
graph = nx.read_edgelist("AlgoProjectDataset.txt", create_using = nx.Graph(), nodetype=int)
```

```python
print (nx.info(graph))
```

```python
sortedNodes = sorted(graph.degree, key=lambda x: x[1], reverse=True)
```

## Qno 1

```python
i = 0
maximum = sortedNodes[i][1]
while (maximum == sortedNodes[i][1]):
    user = sortedNodes[i]
    print "User with ID",user[0],"has maximum Friends."
    print "Total Number of Friends:",user[1]
    print "Friends IDs are........"
    friends = graph.neighbors(user[0])
    for count,f in enumerate(friends):
        print count+1,"User ID",f
    i += 1
```
